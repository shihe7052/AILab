```python
# ==== 可調參數與開關 ====
epochs = 50
z_dim = 100
d_updates_per_g = 2           # 每輪多訓幾步 D，常見 1~5
label_smooth = (0.9, 1.0)     # 真的標籤在 0.9~1.0 隨機
flip_prob = 0.05              # 罕見標籤翻轉，有助避免過擬合
use_r1 = True                 # 啟用 R1 梯度懲罰（對 real）
r1_gamma = 10.0               # 懲罰強度，5~10 常見
use_amp = True                # 啟用混合精度（需要 CUDA）

# ==== 損失函數：用 BCEWithLogitsLoss（D 的輸出 logits 無需 Sigmoid）====
bce = torch.nn.BCEWithLogitsLoss()

scaler = torch.cuda.amp.GradScaler(enabled=use_amp)

# 固定的評估噪聲（可重現的視覺化）
fixed_noise = torch.randn(64, z_dim, device=device)

# 如果你用的是 [-1,1] 歸一化，展示時要反轉；這裡給個小工具
def denorm(x):
    # 假設輸出是 [-1,1]，就把它轉回 [0,1]
    return (x + 1) * 0.5

# 記錄
g_losses, d_losses, g_scores, d_scores = [], [], [], []

G.train()
D.train()

# ?? 建立 Generator 與 Discriminator 模型
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(100, 256),
            nn.ReLU(),
            nn.Linear(256, 784),
            nn.Tanh()
        )
    def forward(self, z):
        return self.model(z).view(-1, 1, 28, 28)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 1),
            # nn.Sigmoid() # Remove Sigmoid here, use BCEWithLogitsLoss
        )
    def forward(self, x):
        # Ensure input is float for linear layers before passing to the model
        return self.model(x.float())


# ?? 初始化模型與優化器
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
G = Generator().to(device)
D = Discriminator().to(device)

g_optimizer = optim.Adam(G.parameters(), lr=0.0001)
d_optimizer = optim.Adam(D.parameters(), lr=0.0001)


for epoch in range(epochs):
    for real_imgs, _ in train_loader:
        real_imgs = real_imgs.to(device, non_blocking=True)
        bsz = real_imgs.size(0)

        # ====== 多步 Discriminator ======
        for _ in range(d_updates_per_g):
            z = torch.randn(bsz, z_dim, device=device)
            with torch.cuda.amp.autocast(enabled=use_amp):
                fake_imgs = G(z).detach()  # 斷開圖以免回傳到 G

                # 一側標籤平滑 + 少量標籤翻轉
                real_targets = torch.empty(bsz, 1, device=device).uniform_(*label_smooth)
                fake_targets = torch.zeros(bsz, 1, device=device)

                if flip_prob > 0:
                    flip_mask_real = (torch.rand_like(real_targets) < flip_prob)
                    flip_mask_fake = (torch.rand_like(fake_targets) < flip_prob)
                    real_targets = torch.where(flip_mask_real, torch.zeros_like(real_targets), real_targets)
                    fake_targets = torch.where(flip_mask_fake, torch.empty_like(fake_targets).uniform_(*label_smooth), fake_targets)

                real_logits = D(real_imgs)     # 不要在 D 裡做 Sigmoid
                fake_logits = D(fake_imgs)

                d_real_loss = bce(real_logits, real_targets)
                d_fake_loss = bce(fake_logits, fake_targets)
                d_loss = d_real_loss + d_fake_loss

                # R1 梯度懲罰（對 real；提升穩定）
                if use_r1:
                    real_imgs.requires_grad_(True)
                    real_logits_for_gp = D(real_imgs)
                    grad_outputs = torch.ones_like(real_logits_for_gp)
                    grads = torch.autograd.grad(
                        outputs=real_logits_for_gp, inputs=real_imgs,
                        grad_outputs=grad_outputs, create_graph=True, retain_graph=True, only_inputs=True
                    )[0]
                    r1_pen = grads.pow(2).view(bsz, -1).sum(dim=1).mean()
                    d_loss = d_loss + (r1_gamma * 0.5) * r1_pen
                    real_imgs = real_imgs.detach()

            d_optimizer.zero_grad(set_to_none=True)
            scaler.scale(d_loss).backward()
            scaler.step(d_optimizer)
            scaler.update()

        # ====== 一步 Generator ======
        z = torch.randn(bsz, z_dim, device=device)
        with torch.cuda.amp.autocast(enabled=use_amp):
            fake_imgs = G(z)
            # 生成器期望被判成真：目標 1（不做平滑）
            g_targets = torch.ones(bsz, 1, device=device)
            g_loss = bce(D(fake_imgs), g_targets)

        g_optimizer.zero_grad(set_to_none=True)
        scaler.scale(g_loss).backward()
        scaler.step(g_optimizer)
        scaler.update()

    # ====== 每個 epoch 的統計 ======
    with torch.no_grad():
        # 「score」用 sigmoid 後的平均比較直觀
        # Explicitly cast Discriminator outputs to float before sigmoid
        d_score_real = torch.sigmoid(D(real_imgs).float()).mean().item()
        d_score_fake = torch.sigmoid(D(fake_imgs).float()).mean().item()
    g_losses.append(g_loss.item())
    d_losses.append(d_loss.item())
    g_scores.append(d_score_fake)  # 生成圖被判為真的置信度
    d_scores.append(d_score_real)  # 真圖被判為真的置信度

    print(f"Epoch {epoch+1}/{epochs} | G: {g_loss.item():.4f} | D: {d_loss.item():.4f} | D(real): {d_score_real:.3f} | D(fake): {d_score_fake:.3f}")

    # ====== 可視化與儲存（每 10 輪）======
    if (epoch + 1) % 10 == 0:
        G.eval()
        with torch.inference_mode():
            show_imgs = G(fixed_noise).detach().cpu()
        G.train()

        # 若通道是 (N,1,28,28)，把它轉成 (N,28,28)
        if show_imgs.ndim == 4 and show_imgs.size(1) == 1:
            show_imgs = show_imgs.squeeze(1)

        import matplotlib.pyplot as plt
        import math
        n = show_imgs.size(0)
        cols = int(math.sqrt(n))
        rows = math.ceil(n / cols)

        plt.figure(figsize=(cols*2, rows*2))
        for i in range(n):
            plt.subplot(rows, cols, i + 1)
            plt.imshow(denorm(show_imgs[i]).clamp(0,1), cmap='gray')
            plt.axis('off')
        plt.suptitle(f'Generated - Epoch {epoch+1}')
        plt.tight_layout()
        plt.savefig(f'generated_epoch_{epoch+1}.png', dpi=150)
        plt.show()

# ====== 繪製曲線 ======
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(g_losses, label='G')
plt.plot(d_losses, label='D')
plt.title('Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()

plt.subplot(1,2,2)
plt.plot(g_scores, label='D(fake)→real prob')
plt.plot(d_scores, label='D(real)→real prob')
plt.title('Scores (sigmoid)')
plt.xlabel('Epoch'); plt.ylabel('Probability'); plt.legend()
plt.tight_layout()
plt.show()
```
